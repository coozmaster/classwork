```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
# PS 5.23
ep = 0.2
x = np.arange(-1.5,1.5+0.01,0.01)
y = (2*np.e)/(np.e**2 - 1.) * np.e**x + 2*np.e * (1 - np.e**2/(np.e**2 - 1))*np.e**(-x) + 1
yinner = np.ones_like(x)
youter = (2*np.e)/(np.e**2 - 1.) * np.e**x + 2*np.e * (1 - np.e**2/(np.e**2 - 1))*np.e**(-x) + 1
plt.figure(0,figsize=(5*np.e,5))
plt.grid('on')
plt.xlabel(r'$x$',fontsize='x-large')
plt.ylabel(r'$y$',fontsize='x-large',rotation='horizontal',horizontalalignment='right')
plt.plot(x,y,'b-')
plt.plot(x,yinner,'g--')
plt.plot(x,youter,'r--')
plt.legend([r'$y\ (approx)$',r'$y_{inner}$',r'$y_{outer}$'],loc='best')
```

```python
%matplotlib inline
# PS 5.35
import numpy as np
import matplotlib.pyplot as plt
d = 0.01
ϵ = 0.2
x = np.arange(-2.,2. +d,d)
y = x + 2 * np.tanh(x/ϵ)
youter = x + 2
yinner = 2 * np.tanh(x/ϵ)
plt.figure(0,figsize=(5*np.e,5))
plt.grid('on')
plt.xlabel(r'$x$',fontsize='x-large')
plt.ylabel(r'$y$',fontsize='x-large',rotation='horizontal',horizontalalignment='right')
plt.plot(x,y,'b-')
plt.plot(x,youter,'g--')
plt.plot(x,yinner,'r--')
plt.legend([r'$y\ (approx)$',r'$y_{outer}$',r'$y_{inner}$'],loc='best')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
# PS 5.54
d = 0.01
t = np.arange(0.,10. + d,d)
ϵ= 0.3
x = (-1/5 * ϵ*t)*np.cos(t) + (-2/5*ϵ*t + 1)*np.sin(t)
xexact = (np.sin(1./20.*np.sqrt(391.)*t))*np.e**(-3./20.*t)
err = (x - xexact)**2

plt.figure(0,figsize=(5*np.e,5))
plt.grid('on')
plt.plot(t,x,'b-')
plt.plot(t,xexact,'g-')
plt.plot(t,err,'r--')
plt.xlabel(r'$t$',fontsize='x-large')
plt.ylabel(r'$x$',fontsize='x-large',rotation='horizontal',horizontalalignment='right')
plt.legend([r'$x\ (approx)$',r'$x\ (exact)$',r'$error^2$'],loc='best')
```
