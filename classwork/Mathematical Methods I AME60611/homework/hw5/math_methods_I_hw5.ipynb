```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
# PS 5.23
ep = 0.2
x = np.arange(-1.5,1.5+0.01,0.01)
y = (2*np.e)/(np.e**2 - 1.) * np.e**x + 2*np.e * (1 - np.e**2/(np.e**2 - 1))*np.e**(-x) + 1
yinner = np.ones_like(x)
youter = (2*np.e)/(np.e**2 - 1.) * np.e**x + 2*np.e * (1 - np.e**2/(np.e**2 - 1))*np.e**(-x) + 1
plt.figure(0,figsize=(5*np.e,5))
plt.grid('on')
plt.xlabel(r'$x$',fontsize='x-large')
plt.ylabel(r'$y$',fontsize='x-large',rotation='horizontal',horizontalalignment='right')
plt.plot(x,y,'b-')
plt.plot(x,yinner,'g--')
plt.plot(x,youter,'r--')
plt.legend([r'$y\ (approx)$',r'$y_{inner}$',r'$y_{outer}$'],loc='best')
```
