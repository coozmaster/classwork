```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


t = np.arange(0.,50.,0.0001)
a = 20. # ft/s**2
m = 1610./32.174 # slug
V = 1000. # ft/s
mdot = 1. # slug/s
F = 0.002 * (a*t)**2. # lbf

theta = np.arccos(1/(mdot*V)*(a*(m - mdot*t) + F))

plt.figure(0,figsize=(5*np.e,5))
plt.grid('on')
plt.plot(t,theta * 180./np.pi,'b-')
plt.xlabel(r'time; $t\ [s]$',fontsize='x-large')
plt.ylabel(r'$\theta\ [deg]$',fontsize='x-large',rotation='horizontal',horizontalalignment='right')
plt.figure(1,figsize=(5*np.e,5))
plt.grid('on')
plt.plot(t,m - mdot*t,'g-')
plt.xlabel(r'time; $t\ [s]$',fontsize='x-large')
plt.ylabel(r'Mass of Cart; $m_{cart} [slug]$',rotation='horizontal',horizontalalignment='right',fontsize='x-large')
hold = t[~np.isnan(theta)]
print(hold)
```
