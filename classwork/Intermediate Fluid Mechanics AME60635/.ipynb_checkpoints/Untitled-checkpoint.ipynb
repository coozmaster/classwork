```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
x0 = 20
y0 = 20
t0 = 0

ypos = np.arange(-100,100,10)
xpos = np.arange(-100,100,10)

x,y = np.meshgrid(xpos,ypos)
wind_x0 = 4 + 0.1*ypos
wind_y0 = 2 * np.ones_like(ypos)
plt.figure(0,figsize=(5*np.e,5))
plt.quiver(x,y,wind_x0,wind_y0,color='g'); plt.xlabel('X'); plt.ylabel('Y')
plt.title('Velocity Field')

dt = 0.001
ti = np.arange(0,10. + dt,dt)
yi = np.zeros_like(ti)
xi = np.zeros_like(ti)
uxi = np.zeros_like(ti)
uyi = np.zeros_like(ti)

# Pathline
for i in range(len(ti)):
   if( i == 0):
       yi[0] = 20
       xi[0] = 20
       uxi[0] = 4+0.1*yi[0]
       uyi[0] = 2
   else:
       xi[i] = uxi[i-1] * (ti[i] - ti[i-1]) + xi[i-1]
       yi[i] = uyi[i-1] * (ti[i] - ti[i-1]) + yi[i-1]
       uxi[i] = (xi[i] - xi[i-1])/(ti[i] - ti[i-1])
       uyi[i] = (yi[i] - yi[i-1])/(ti[i] - ti[i-1])

plt.figure(1,figsize=(5*np.e,5))
plt.plot(xi,yi,'-b'); plt.xlabel('X'); plt.ylabel('Y'); plt.legend(['Pathline'],loc='best')

```

```python
# Now for the frisbee
m = 175
cd = 0.1
rho = 1.2041 
A = 0.8 # m**2

dt = 0.001
ti = np.arange(0,2. + dt,dt)
yi = np.zeros_like(ti)
xi = np.zeros_like(ti)
uxi = np.zeros_like(ti)
uyi = np.zeros_like(ti)
wind_xi = np.zeros_like(ti)
wind_yi = np.zeros_like(ti)

# Pathline
for i in range(len(ti)):
   if( i == 0):
       yi[0] = 20
       xi[0] = 20
       wind_xi[0] = 4+0.1*yi[0]
       wind_yi[0] = 2
       uyi[0] = 20
       uxi[0] = 0
   else:
       xi[i] = uxi[i-1] * (ti[i] - ti[i-1]) + xi[i-1]
       yi[i] = uyi[i-1] * (ti[i] - ti[i-1]) + yi[i-1]
       uxi[i] = 1/m*(0.5*rho*(wind_xi[i-1] - uxi[i-1])**2. *cd *A) * (ti[i]-ti[i-1]) + uxi[i-1]
       uyi[i] = 1/m*(0.5*rho*(wind_yi[i-1] - uyi[i-1])**2. *cd *A) * (ti[i]-ti[i-1]) + uyi[i-1]
       wind_xi[i] = 4+0.1*yi[i] -0.2*ti[i]
       wind_yi[i] = 2

print("Xpos at t==2 ",xi[ti[:]==2]," and YPos at t==2 ",yi[ti[:]==2])
plt.figure(0,figsize=(5*np.e,5))
plt.plot(xi,yi)
plt.legend(['Frisbee Path'],loc='best'); plt.xlabel('X'); plt.ylabel('Y')
```

The $(x,y)$ at $t = 2~s$ was found to be 20.0288 for X and 60.1794 for Y.
As the magnitude of the wind increases in the X-direction you can see the momentum in the X-direction begin to accumulate dramatically. This causes the highly curved shape.

Differences between the pathline at $t=0$ for a particle are due to different factors. One being the initial velocity in the Y direction before sufficient time the frisbee path follows the velocity field due to the presence of the drag force on the frisbee.
