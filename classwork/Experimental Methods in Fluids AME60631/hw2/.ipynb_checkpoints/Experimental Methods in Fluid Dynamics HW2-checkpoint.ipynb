Example from class:

Consider the frequency response of system for input

\begin{equation}
q_i = sin(2t) + 0.3 sin(20t)
\end{equation}

Where the former term is the low frequency component and the latter is the high frequency component.

### Variable Declaration

Various variables defined here:

$\tau = 0.2~s$; Time constant

$A_1 = 1$; Amplitude (or gain) for low freq term

$A_2 = 0.3$; Amplitude (or gain) for high freq term

Consider 1st order system with $\tau = 0.2~s$

Now,

\begin{equation}
q_i = A_1 sin(\omega_1 t + \phi_1) + A_2 sin(\omega_2 t + \phi_2)
\end{equation}

since it is a linear system we may use linear superposition to find the output $q_o$ as,

\begin{equation}
q_o = \left|\frac{q_o(i\omega_1)}{q_i} \right| A_1 sin(\omega_1 t + \phi_1) + \left|\frac{q_o(i\omega_2)}{q_i}\right| A_2 sin(\omega_2 t + \phi_2)
\end{equation}

Then evaluate the sinusoidal transfer function at the 2 frequencies of interest for $\tau = 0.2~s$.

\begin{equation}
\frac{q_o(i\omega_1)}{q_i}|_{\omega=2} = \frac{K}{\sqrt{1 + \omega_1^2 \tau^2}}
\end{equation}

Where $K$ is the scaling factor due to gain

\begin{equation}
\phi_1 = atan(-\omega_1 \tau)
\end{equation}

\begin{equation}
\frac{q_o(i\omega_2)}{q_i}|_{\omega=20} = \frac{K}{\sqrt{1 + \omega_2^2 \tau^2}}
\end{equation}

\begin{equation}
\phi_2 = atan(-\omega_2 \tau)
\end{equation}

Plugging in the variables yields...

\begin{equation}
\frac{q_o(i\omega_1)}{q_i}|_{\omega=2} = 0.93 K
\end{equation}

\begin{equation}
\phi_1 = -21.8^o
\end{equation}

\begin{equation}
\frac{q_o(i\omega_2)}{q_i}|_{\omega=20} = (0.24) K
\end{equation}

\begin{equation}
\phi_2 = -76^o
\end{equation}

Then returning back to the original equation and solving for $\frac{q_o}{K}$...

\begin{equation}
\frac{q_o}{K} = 0.93 ~sin(2t - 0.38048) + 0.072~sin(20t - 1.32645)
\end{equation}

So we can see that the low frequency component as a slight attenuation of 0.93 while the high frequency content is severally attenuated with 0.072

```python
## Experimental Methods in Fluid Dynamics
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

A1 = 1.
A2 = 0.3
ω1 = 2.
ω2 = 20.
τ = 0.2

qo1_qi1 = 1/np.sqrt(1 + ω1**2. * τ**2.)
phi1 = np.arctan2(-ω1 * τ,1.)
qo2_qi2 = 1/np.sqrt(1 + ω2**2. * τ**2.)
phi2 = np.arctan2(-ω2 * τ,1.)

print(r'qo1_qi1 ',qo1_qi1)
print(r'qo2_qi2 ',qo2_qi2)
print(r'phi1 ',phi1)
print(r'phi2 ',phi2)

t = np.arange(0,np.pi,10**(-3.))
t = np.arange(0,np.pi,10**(-3.))
#qo1_qi1 = 1/np.sqrt( 1 + t**2.)
#qo2_qi2 = 1/np.sqrt( 1 + t**2.)
#phi1 = np.arctan2(-t,1)
#phi2 = np.arctan2(-t,1)
qo_k = A1 * qo1_qi1 * np.sin(ω1 * t + phi1) + A2 * qo2_qi2 * np.sin(ω2 * t + phi2)
qi   = A1 * np.sin(ω1*t)  + A2 *np.sin(ω2*t)

plt.figure(0,figsize=(5*np.e,5))
plt.xlabel(r'$t$',fontsize='x-large')
plt.ylabel(r'$Amplitude$',fontsize='x-large')
#plt.yscale('log')
plt.grid('on')
plt.plot(t,qo_k,'-b',t,qi,'g--')
plt.legend([r'$q_o/k$',r'$q_i$'],loc='best')
```

```python
## Experimental Methods in Fluid Dynamics
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

A1 = 1.
A2 = 0.3
ω1 = 2.
ω2 = 20.
τ = 0.002

qo1_qi1 = 1/np.sqrt(1 + ω1**2. * τ**2.)
ϕ1 = np.arctan2(-ω1 * τ,1.)
qo2_qi2 = 1/np.sqrt(1 + ω2**2. * τ**2.)
ϕ2 = np.arctan2(-ω2 * τ,1.)

ωτ = np.arange(0,np.e*np.pi,10**(-3.))

qo_kqi = 1/np.sqrt( 1 + ωτ**2.)
ϕ = np.arctan2(-ωτ,1)
plt.figure(0,figsize=(5*np.e,5))
plt.xlabel(r'$\omega\tau$',fontsize=30)
plt.ylabel(r'$\frac{q_o}{kq_i}$',fontsize=30,rotation='horizontal',horizontalalignment='right')
plt.grid('on')
plt.plot(ωτ,qo_kqi,'-b')

plt.figure(1,figsize=(5*np.e,5))
plt.xlabel(r'$\omega\tau$',fontsize=30)
plt.ylabel(r'$\phi~[rad]$',fontsize=30,rotation='horizontal',horizontalalignment='right')
plt.grid('on')
plt.plot(ωτ,ϕ,'-g')
```
